```python
from memo import memo
import jax
import jax.numpy as np
from enum import IntEnum
```

```python
C = np.array([0,1,2]) #three doors to choose from
H = np.array([0,1,2]) #three possible hints that the MC could give


@memo
def Part[c1: C, h: H, c2:C](): #c1: first choice, h: hint, c2: final choice
    participant: knows(c1)
    participant: thinks[
         world: chooses(c_true in C, wpp = 1),
         emcee: knows(world.c_true),
         emcee: knows(c1),
         emcee: chooses(h_said in H, wpp = (h_said != world.c_true)*(h_said != c1)*1)]
    participant: observes[emcee.h_said] is h
    participant: chooses(c_final in C, wpp = Pr[emcee[world.c_true] == c_final])
    return Pr[participant.c_final == c2]
```
